### Task 1: For each category, list the 5 most used ingredients:

1. Define the categories we are interested in:

```python
category = {
    "Protein":["egg", "chicken", "beef", "pork", "lamb", "fish", "shrimp"],
    "Vegetable": ["onion", "tomato","bell_pepper", "potato", "celery", "bean", "mushroom"],
    "Herbs & Spice": ["black_pepper", "garlic", "cayenne", "cumin", "ginger", "cilantro", "cinnamon"],
    "Condiment": ["vinegar", "lemon", "soy_sauce", "mustard", "chicken_broth", "white_wine", "red_wine"],
    "Grain & Product":["wheat_flour", "rice", "bread", "corn", "starch"],
    "Oil & Dairy":["vegetable_oil", "olive_oil", "sesame_oil", "butter", "milk", "cream"]
}
```

```python
cuisine_index = all_cuisines_all_ing["Country"].values
```

```python
columns = []
ingredient_pct = []
for i, cat in enumerate(category):
    ingredient_df = all_cuisines_all_ing[category[cat]]
    ingredient_dict = ingredient_df.to_dict(orient="split")
    large_dict = []
    for j,row in enumerate(ingredient_dict["data"]):
        medium_dict = []
        for k,ingredient in enumerate(ingredient_dict["columns"]):
            small_dict = {"ingredient":ingredient, "num":row[k]}
            medium_dict.append(small_dict)
        large_dict.append(medium_dict)
    columns.append(large_dict)
    ingredient_pct.append(np.sum(ingredient_df.values, axis=1))

category_pct = np.transpose(np.array(ingredient_pct)*1./np.sum(ingredient_pct, axis=0))
```

```python
cuisine_ingredient_df = pd.DataFrame(data=np.array(columns).T, columns=category, index=cuisine_index)
# pandas.DataFrame(data=None, index=None, columns=None, dtype=None, copy=False)
```

```python
cuisine_ingredient_df
```

```python
category_pct
```

```python
category_index = category.keys()
print category_index
category_pct_list = []
for i,pct_list in enumerate(category_pct):
    pct_dict_list=[]
    for j,pct in enumerate(pct_list):
        pct_dict = {"category": category_index[j], "percent": pct}
        pct_dict_list.append(pct_dict)
    category_pct_list.append(pct_dict_list)
```

```python
category_pct_list
```

```python
cuisine_ingredient_df["category_pct"] = category_pct_list
```
